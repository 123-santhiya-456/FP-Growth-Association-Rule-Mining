```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df=pd.read_csv("/content/drive/MyDrive/Market_Basket_Optimisation.csv",header=None) #header=None
df.head()
```

```python
# Convert transactions to list
transactions = df.values.tolist()
```

```python
transactions_cleaned = []
for transaction in transactions:
    transactions_cleaned.append([item for item in transaction if str(item) != 'nan'])

te = TransactionEncoder()
te_ary = te.fit(transactions_cleaned).transform(transactions_cleaned)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# FP-Growth
frequent_itemsets = fpgrowth(
    df_encoded,
    min_support=0.02,
    use_colnames=True
)
```

```python
# Association Rules
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)
```
